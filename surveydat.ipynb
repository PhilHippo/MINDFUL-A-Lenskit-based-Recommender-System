```python
import pandas as pd
```

```python
news = pd.read_csv('small_test_data/news.tsv',header=None,sep='\t')
```

```python
news.columns=['News ID',
"Category",
"SubCategory",
"Title",
"Abstract",
"URL",
"Title Entities",
"Abstract Entities "]
```

```python
# Create a DataFrame with unique subcategories
unique_subcategories = news['SubCategory'].unique()
subcategory_rows = pd.DataFrame(columns=news.columns)
# Ensure you have at least 20 unique subcategories
if len(unique_subcategories) < 20:
    print("Not enough unique subcategories.")
else:
    # Randomly shuffle the unique subcategories
    import random
    random.shuffle(unique_subcategories)
    # Filter the original DataFrame to select rows with the chosen subcategories
    for subcategory in unique_subcategories:
        subcategory_data = news[news['SubCategory'] == subcategory]
        if subcategory_data.shape[0] >= 2:
            subcategory_data = subcategory_data.head(2)
            subcategory_rows = pd.concat([subcategory_rows, subcategory_data])
        if subcategory_rows.shape[0] >= 40:
            break


# Display the selected rows
subcategory_rows
```
